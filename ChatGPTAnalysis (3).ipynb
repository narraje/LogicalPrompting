```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/ChatGPTFinal.xlsx'
excel_data = pd.ExcelFile(file_path)

sheet1_data = excel_data.parse('Step1')

print(sheet1_data.head())
average_scores = sheet1_data[['Accuracy', 'Completeness', 'Logical Soundess',
                              'ACL2s Adherence', 'Efficiency', 'Hallucinations']].mean()

print(average_scores)
sheet1_data['Aggregate Score'] = sheet1_data[['Accuracy', 'Completeness', 'Logical Soundess',
                                              'ACL2s Adherence', 'Efficiency', 'Hallucinations']].mean(axis=1)
sheet1_data['Rounded Aggregate Score'] = sheet1_data['Aggregate Score'].round()

plt.figure(figsize=(10, 6))
average_scores.plot(kind='bar', rot=45)
plt.title('Average Scores Across Criteria')
plt.ylabel('Average Score')
plt.xlabel('Criteria')
plt.ylim(0, 5)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sheet1_data['Rounded Aggregate Score'].value_counts().sort_index().plot(kind='bar', rot=0)
plt.title('Rounded Aggregate Score Distribution')
plt.xlabel('Rounded Aggregate Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

sheet2_data = excel_data.parse('Step2')

print(sheet2_data.head())
average_scores = sheet2_data[['Accuracy', 'Completeness', 'Logical Soundess',
                              'ACL2s Adherence', 'Efficiency', 'Hallucinations']].mean()

print(average_scores)
sheet2_data['Aggregate Score'] = sheet2_data[['Accuracy', 'Completeness', 'Logical Soundess',
                                              'ACL2s Adherence', 'Efficiency', 'Hallucinations']].mean(axis=1)
sheet2_data['Rounded Aggregate Score'] = sheet2_data['Aggregate Score'].round()

plt.figure(figsize=(10, 6))
average_scores.plot(kind='bar', rot=45)
plt.title('Average Scores Across Criteria')
plt.ylabel('Average Score')
plt.xlabel('Criteria')
plt.ylim(0, 5)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sheet2_data['Rounded Aggregate Score'].value_counts().sort_index().plot(kind='bar', rot=0)
plt.title('Rounded Aggregate Score Distribution')
plt.xlabel('Rounded Aggregate Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

sheet3_data = excel_data.parse('Step3')

print(sheet3_data.head())
average_scores = sheet3_data[['Accuracy', 'Completeness', 'Logical Soundess',
                              'ACL2s Adherence', 'Efficiency', 'Hallucinations']].mean()

print(average_scores)
sheet3_data['Aggregate Score'] = sheet3_data[['Accuracy', 'Completeness', 'Logical Soundess',
                                              'ACL2s Adherence', 'Efficiency', 'Hallucinations']].mean(axis=1)
sheet3_data['Rounded Aggregate Score'] = sheet3_data['Aggregate Score'].round()

plt.figure(figsize=(10, 6))
average_scores.plot(kind='bar', rot=45)
plt.title('Average Scores Across Criteria')
plt.ylabel('Average Score')
plt.xlabel('Criteria')
plt.ylim(0, 5)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sheet3_data['Rounded Aggregate Score'].value_counts().sort_index().plot(kind='bar', rot=0)
plt.title('Rounded Aggregate Score Distribution')
plt.xlabel('Rounded Aggregate Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

sheet4_data = excel_data.parse('PromptSubset')

print(sheet4_data.head())
average_scores = sheet4_data[['Accuracy', 'Completeness', 'Logical Soundess',
                              'ACL2s Adherence', 'Efficiency', 'Hallucinations']].mean()

print(average_scores)
sheet4_data['Aggregate Score'] = sheet4_data[['Accuracy', 'Completeness', 'Logical Soundess',
                                              'ACL2s Adherence', 'Efficiency', 'Hallucinations']].mean(axis=1)
sheet4_data['Rounded Aggregate Score'] = sheet4_data['Aggregate Score'].round()

plt.figure(figsize=(10, 6))
average_scores.plot(kind='bar', rot=45)
plt.title('Average Scores Across Criteria')
plt.ylabel('Average Score')
plt.xlabel('Criteria')
plt.ylim(0, 5)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sheet4_data['Rounded Aggregate Score'].value_counts().sort_index().plot(kind='bar', rot=0)
plt.title('Rounded Aggregate Score Distribution')
plt.xlabel('Rounded Aggregate Score')
plt.ylabel('Frequency')
plt.xlim(0,5)
plt.tight_layout()
plt.show()
```
