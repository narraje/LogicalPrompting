```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'CHATGPT.xlsx'
excel_data = pd.ExcelFile(file_path)

print(excel_data.sheet_names)

sheet1_data = excel_data.parse('Sheet1')
print(sheet1_data.head())

average_scores = sheet1_data.mean(numeric_only=True)
print(average_scores)


average_scores.plot(kind='bar', title='Average Scores Across Criteria', ylabel='Average Score', xlabel='Criteria')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

sheet1_data['Aggregate Score'] = sheet1_data.mean(axis=1, numeric_only=True)

sheet1_data['Rounded Aggregate Score'] = sheet1_data['Aggregate Score'].round()

rounded_aggregate_score_distribution = sheet1_data['Rounded Aggregate Score'].value_counts().sort_index()

rounded_aggregate_score_distribution.plot(kind='bar', title='Rounded Aggregate Score Distribution', xlabel='Rounded Aggregate Score', ylabel='Frequency')
plt.tight_layout()
plt.show()
```
