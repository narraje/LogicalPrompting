```python
import pandas as pd
import matplotlib.pyplot as plt

gemini_file_path = '/content/Geminii.xlsx'
gemini_data = pd.ExcelFile(gemini_file_path)

gemini_sheet1_data = gemini_data.parse('Sheet1')

average_scores = gemini_sheet1_data[['Accuracy', 'Comleteness', 'Logical Soundess',
                                     'ACL2s Adherence', 'Efficiency', 'Hallucinations']].mean()
gemini_sheet1_data['Aggregate Score'] = gemini_sheet1_data[['Accuracy', 'Comleteness', 'Logical Soundess',
                                                            'ACL2s Adherence', 'Efficiency', 'Hallucinations']].mean(axis=1)
gemini_sheet1_data['Rounded Aggregate Score'] = gemini_sheet1_data['Aggregate Score'].round()

plt.figure(figsize=(10, 6))
average_scores.plot(kind='bar', rot=45)
plt.title('Average Scores Across Criteria (Gemini Results)')
plt.ylabel('Average Score')
plt.xlabel('Criteria')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
gemini_sheet1_data['Rounded Aggregate Score'].value_counts().sort_index().plot(kind='bar', rot=0)
plt.title('Rounded Aggregate Score Distribution (Gemini Results)')
plt.xlabel('Rounded Aggregate Score')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```
