```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data structure for the three models
# Assuming the structure of data for each model is consistent and columns are the same
data = {
    "Model": ["ChatGPT", "Gemini", "CoPilot"],
    "Accuracy": [3.320755, 2.641509, 3.185714],
    "Completeness": [4.000000, 3.415094, 3.385714],
    "Logical Soundness": [2.800000,  2.600000, 3.242857],
    "ACL2s Adherence": [2.384615, 1.735849, 2.971429],
    "Efficiency": [3.415094, 3.433962, 3.100000],
    "Hallucinations": [3.773585, 3.471698, 3.200000]
}

# Convert the dictionary into a DataFrame
cross_model_data = pd.DataFrame(data)
cross_model_data.set_index("Model", inplace=True)

# Plotting a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cross_model_data, annot=True, cmap="coolwarm", cbar=True, fmt=".1f", linewidths=0.5)
plt.title("Performance Heatmap Across Models")
plt.ylabel("Model")
plt.xlabel("Criteria")
plt.tight_layout()
plt.show()
```
