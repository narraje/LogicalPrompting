```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import f_oneway

data = {
    "Stage": ["Collective Prompting", "Lecture Notes", "Collective Prompting 2", "Extended Prompting"],
    "ChatGPT-Accuracy": [3.320755, 3.452830, 2.981132, 3.000000],
    "Gemini-Accuracy": [2.641509, 2.603774, None, 3.545455],
    "CoPilot-Accuracy": [3.185714, None, 3.342857, 3.620689655],
    "ChatGPT-Completeness": [4.000000, 3.886792, 3.358491, 4.222222],
    "Gemini-Completeness": [3.415094, 3.000000, None, 4.000000],
    "CoPilot-Completeness": [3.385714, None, 3.385714, 3.586206897],
    "ChatGPT-Logical_Soundness": [2.800000, 2.866667, 2.333333, 2.500000],
    "Gemini-Logical_Soundness": [2.600000, 1.533333, None, 2.818182],
    "CoPilot-Logical_Soundness": [3.242857, None, 3.385714, 3.620689655],
    "ChatGPT-ACL2s": [2.384615, 3.320755, 2.622642, 3.833333],
    "Gemini-ACL2s": [1.735849, 2.113208, None, 2.909091],
    "CoPilot-ACL2s": [2.971429, None, 2.614286, 3.000000],
    "ChatGPT-Efficiency": [3.415094, 2.792453, 3.169811, 2.722222],
    "Gemini-Efficiency": [3.433962, 2.716981, None, 3.454545],
    "CoPilot-Efficiency": [3.100000, None, 3.214286, 3.379310345],
    "ChatGPT-Hallucinations": [3.773585, 3.566038, 3.339623, 3.944444],
    "Gemini-Hallucinations": [3.471698, 3.037736, None, 3.909091],
    "CoPilot-Hallucinations": [3.200000, None, 3.314286, 3.413793103],
}

df = pd.DataFrame(data)

df_melted = df.melt(id_vars=["Stage"], var_name="Model_Criterion", value_name="Score")

df_melted[['Model', 'Criterion']] = df_melted['Model_Criterion'].str.split('-', n=1, expand=True)

df_melted = df_melted.drop(columns=["Model_Criterion"])

criteria = df_melted['Criterion'].unique()
for criterion in criteria:
    plt.figure(figsize=(12, 6))
    subset = df_melted[df_melted['Criterion'] == criterion]
    sns.lineplot(data=subset, x="Stage", y="Score", hue="Model", marker="o")
    plt.title(f"{criterion} Scores Across Stages")
    plt.ylabel("Score")
    plt.ylim(0, 5)
    plt.xlabel("Stage")
    plt.legend(title="Model")
    plt.tight_layout()
    plt.show()

# Statistical analysis for each criterion
print("\nStatistical Analysis:")
for criterion in criteria:
    print(f"\n--- {criterion} ---")
    subset = df_melted[df_melted['Criterion'] == criterion].pivot(index="Stage", columns="Model", values="Score")
    subset = subset.dropna(axis=0, how="any")
    if len(subset) > 1:
        for model in subset.columns:
            scores = subset[model].values
            stat, p_value = f_oneway(*[scores for stage in subset.index])
            print(f"{model}: F-Statistic = {stat:.4f}, P-Value = {p_value:.4f}")

from scipy.stats import friedmanchisquare, wilcoxon
import numpy as np

# Function to apply non-parametric tests
def run_nonparametric_tests(df_melted, test="friedman"):
    criteria = df_melted['Criterion'].unique()
    results = []

    for criterion in criteria:
        subset = df_melted[df_melted['Criterion'] == criterion]
        pivoted = subset.pivot(index="Stage", columns="Model", values="Score")

        print(f"\n--- {criterion} ---")
        for model in pivoted.columns:
            scores = pivoted[model].dropna().values

            if len(scores) > 1:
                if test == "friedman" and len(pivoted.index) > 2:
                    # Friedman Test
                    stat, p_value = friedmanchisquare(*pivoted.dropna().values.T)
                    results.append((criterion, model, "Friedman", stat, p_value))
                    print(f"{model}: Friedman Test - Chi-Square = {stat:.4f}, P-Value = {p_value:.4f}")
                elif test == "wilcoxon":
                    # Pairwise Wilcoxon Test (e.g., Stage 1 vs Stage 2)
                    for i in range(len(pivoted.index) - 1):
                        try:
                            stat, p_value = wilcoxon(
                                pivoted.iloc[i, :].dropna(), pivoted.iloc[i + 1, :].dropna()
                            )
                            results.append((criterion, model, f"Wilcoxon Stage {i+1} vs {i+2}", stat, p_value))
                            print(f"{model}: Wilcoxon Test (Stage {i+1} vs {i+2}) - Stat = {stat:.4f}, P-Value = {p_value:.4f}")
                        except ValueError:
                            print(f"{model}: Insufficient data for Wilcoxon Test between Stage {i+1} and Stage {i+2}")

    return results

print("\n--- Friedman Test Results ---")
friedman_results = run_nonparametric_tests(df_melted, test="friedman")

print("\n--- Wilcoxon Test Results ---")
wilcoxon_results = run_nonparametric_tests(df_melted, test="wilcoxon")


```
